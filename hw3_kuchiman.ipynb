**2.41** Why are the boxplot whiskers chosen with the factor of 1.5? Why not some other factor? You can see the results of other choices by setting the range=argument. Use x=rnorm(1000) for your data. Try values of 0.5, 1, 1.5, and 2 to see which shows the tails of the distribution best. (This random sample should not have a heavy tail or a light tail, meaning it will usually have a handful of points beyond the whiskers in a sample of this size.) 

Boxplot whiskers are chosen with the factor of 1.5 IQR as that is a standard way of identifying outliers in the data set. It is the standard because if the factor is too low or too high, the outlier tend to affect the distribution. Thus, the standard was defined as a reproducible and effective way of identifying the majority of the distribution while optimally removing outliers.

```r
x = rnorm(1000)
par(mfrow=c(2,2))
boxplot(x, horizontal = TRUE, range = 0.5, main = '0.5 * IQR')
boxplot(x, horizontal = TRUE, range = 1, main = '1  *  IQR')
boxplot(x, horizontal = TRUE, range = 1.5, main = '1.5  *  IQR - (Normal)')
boxplot(x, horizontal = TRUE, range = 2, main = '2  *  IQR')
```

**2.45** Sometimes a data set is so skewed that it can help if we transform the data prior to looking at it. A common transformation for long-tailed data sets is to take the logarithm of the data. For example, the exec.pay (UsingR) data set is highly skewed. Look at histograms before and after taking a logarithmic transform. Which is better at showing the data and why? (You can transform with the command log (1+exec.pay, 10).) Find the median and the mean for the transformed data. How do they correspond to the median and mean of the untransformed data? 

```r
pay <- exec.pay
par(mfrow=c(2,1))
hist(pay, main = 'Exec Pay histogram')
hist(log(pay+1, 10), main = 'Log of Exec Pay Histogram')
```

The data is better in the logarithmic transformation as we are able to see the frequency of numerous points with a spread-out frequency. The non-log graph has is heavily edge distributed skewing the data.

```r
summary(log(1+pay, 10))[3:4]
summary(pay)[3:4]
```

The mean and median have a proportional difference between the log data and the original data. Setting the log values to the power of 10 makes them the original values.
